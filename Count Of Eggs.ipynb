```python
import cv2
import os 
import supervision as sv
from  ultralytics import YOLO
import numpy as np
os.chdir(r'D:\Motion-Detection-and-Tracking-Using-Opencv-main\YOLO\yolov8')
```

```python
model=YOLO('best egg.pt')
```

```python
w,h=640,480
cap=cv2.VideoCapture('egg.mp4')
track=sv.ByteTrack()
zone=np.array([[230, 34],[245, 34],[245, 406],[230, 406]])
count_set=set()
font=cv2.FONT_HERSHEY_COMPLEX
forcc=cv2.VideoWriter_fourcc(*'XVID')
out=cv2.VideoWriter('Count Of Eggs.mp4',forcc,20.0,(w,h))
while cap.isOpened():
    _,frame=cap.read()
    frame=cv2.resize(frame,(w,h))
    result=model(frame)[0]
    detections=sv.Detections.from_yolov8(result)
    detections=track.update_with_detections(detections)
    cv2.polylines(frame,[zone],True,(255,0,0),2)
    if detections.tracker_id is not None :

        for box ,idx in zip (detections.xyxy,detections.tracker_id):
            x,y,w2,h2=[int(box[0]),int(box[1]),int(box[2]),int(box[3])]


            cx,cy=int((x+w2)/2),int((y+h2)/2)
            cv2.putText(frame,str(idx),(x,y-10),font,.5,(0,0,0),2)

            cv2.rectangle(frame,(x,y),(w2,h2),(0,0,255),1)
            res=cv2.pointPolygonTest(zone,(cx,cy),False)
            if res >0 :
                count_set.add(idx)
                cv2.rectangle(frame,(x,y),(w2,h2),(0,255,0),1)
                cv2.circle(frame,(cx,cy),2,(0,255,0),2)

    cv2.putText(frame,"Count Of Eggs Is"+ " "+str(len(count_set)),(20,20),font,.6,(10,200,105),2)
    out.write(frame)
    cv2.imshow('frame',frame)
    if cv2.waitKey(1)==27:
        break
out.release()
cap.release()
cv2.destroyAllWindows()
```
